```python
# Using Scipy and Numpy for Plotting (Optimizing and Minimizing)
# Skecting line equation
# f = ax+ b a simple linear equation

import numpy as np
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
# curve fit is best fit method
# in the example we first generate data from a known function with noise and then 
# fit the noise data with curvefit method

def func(x, a,b):
    return a*x + b

x= np.linspace(0,10,100)
y = func(x,1,2)

# Adding noise
yn = y + 0.9*np.random.normal(size=len(x))
popt, pcov = curve_fit(func, x , yn)

print(popt)
plt.figure(figsize=(7.5,5))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,yn,'bo', label='Noise') # kx
plt.plot(x,y, label='Original Function', c='r')
plt.plot(x,func(x,popt[0],popt[1]), label='Best Fit', lw=3)
plt.legend(loc=8)
plt.show()
```

```python
# y = a(x+b)2 + c
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def func2(x,a,b,c):
    y = a*(x**2)+ x*b + c
    return y

x = np.linspace(1,10)
y= func2(x,1,10,2)
yn = y+ 0.9*np.random.normal(size=len(x))

popt, pcov= curve_fit(func2,x,yn)
plt.figure(figsize=(6,5))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,yn,'bo', label='Noise') # kx
plt.plot(x,y, label='Original Function', c='r')
plt.plot(x,func2(x,popt[0],popt[1],popt[2]), label='Best Fit', lw=3)
plt.legend(loc=8)
plt.show()
```

```python
# implementing nonlinear Gaussian or Normal Distribution

def func2(x,a,b,c):
    return a*np.exp(-(x-b)**2/(2*c**2))
x = np.linspace(0,10,100)
y = func2(x,1,5,2)

yn = y + 0.2*np.random.normal(size=len(x))

popt, pcov= curve_fit(func2, x,yn)
plt.figure(figsize=(6,4))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.plot(x,yn, 'bo')
plt.plot(x, func2(x,popt[0],popt[1],popt[2]), label='Best Fit', lw=3)
plt.plot(x,y,c='g',label='Original Function')
plt.legend(loc=8)
plt.show()
```

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
def findIntersection(func1, func2,x0):
    return fsolve(lambda x : func1(x)-func2(x), x0)
funky = lambda x : np.cos(x/5) * np.sin(x/2)
line = lambda x : 0.01 * x - 0.5
x = np.linspace(0,45,10000)
x0 = [x for x in range(15, 50 , 5)]
result = findIntersection(funky, line, x0)

plt.figure(figsize=(10,6))
plt.plot(x, line(x), label="Line")
plt.plot(x,funky(x),label="Funky")
plt.plot(result,funky(result),'ro',label="Point Intersection")
plt.legend(loc=1)
plt.show()
```
